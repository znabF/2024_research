```python
# import diart.operators as dops
# import numpy as np
# import rich
# import rx.operators as ops
# import whisper_timestamped as whisper
# from diart import SpeakerDiarization, SpeakerDiarizationConfig
# from diart.sources import MicrophoneAudioSource
# from pyannote.core import Annotation, SlidingWindowFeature, SlidingWindow, Segment

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, Activation
import tensorflow.keras as keras
import numpy as np # brings functionality of arrays data structure to python, used a lot in machine learning
# import whisper
import cv2
import wave,struct
import struct

# all libs for audio/whisper transcription
import sounddevice as sd # recording compatibility
import wavio as wv # audio file compatibilty
import datetime # to ensure it is live recording by keeping track of time
import config
import os, glob

# from pvrecorder import PvRecorder # instance object of the recorder

# all libs for whisper transcription
# import logging
# import os
# import sys
# import traceback
# from contextlib import contextmanager
# import pyannote

import whisper

        

# cv_model = tf.keras.models.load_model("cv_model.h5")

# curating a model on pre-trained weights (individually trained by me)
cv_model = Sequential()

# activation function is relu for all layers except for last layer (coded), it is relu because relu knows which neurons can activate and which can not
# relu returns input if greater than 0, this input is passed across the network multiple times so network can remember patterns after convolution/pooling layers applied
# mutliple convolution layers required to understand the face image in depth

# convolution layer of 32 filters performing convolution operation of each 3x3 portion of image input
# therefore multiple outputs since more than 1 filter so a feature map will occur
# padding = 'same' means input_shape = output_shape, helps since we can add as much as layers as we want and not to worry about sizing
# kernel_initializer is setting up how weights should be randomly initialized, he_normal is method to initialize weights
cv_model.add(Conv2D(32, (3, 3), input_shape=(48, 48, 1), padding = 'same', kernel_initializer = "he_normal"))

# activation function is relu for all layers except for last layer (coded), it is relu because relu knows which neurons can activate and which can not
cv_model.add(Activation("relu"))

# making sure pixels on same scale after convolution operation
cv_model.add(BatchNormalization())

# max pooling layer, for each 2x2 section of input gets the highest pixel value and stores it in a different feature, another way to size down the image
cv_model.add(MaxPooling2D(pool_size = (2, 2)))

# convolution layer of 32  filters with kernel size 3x3
cv_model.add(Conv2D(32, (3, 3), input_shape=(48, 48, 1), padding = 'same', kernel_initializer = "he_normal"))

# activation function is relu for all layers except for last layer (coded), it is relu because relu knows which neurons can activate and which can not
cv_model.add(Activation("relu"))

# batch normalization makes sure features on same scale, for better accuracy, since the training takes batches at a time
cv_model.add(BatchNormalization())

# max pooling layer
cv_model.add(MaxPooling2D(pool_size = (2, 2)))

# prevents overfitting when neurons get dropped less patterns to remember so it is generalized
cv_model.add(Dropout(0.1))

# convolution layer of 64, complexities of layers good for this type of project
cv_model.add(Conv2D(64, (3 , 3), padding = 'same', kernel_initializer = "he_normal"))

# activation is relu, how data propogates unlinearly
cv_model.add(Activation("relu"))

# batch normalization makes sure features on same scale, for better accuracy, since the training takes batches at a time
cv_model.add(BatchNormalization())

# pooling operations always go with convolution layer
cv_model.add(MaxPooling2D(pool_size = (2, 2)))

# convolution layer of 64 filters performing convolution operation of size 3x3
# therefore multiple outputs since more than one filter so a feature map will occur
cv_model.add(Conv2D(64, (3, 3), padding = 'same', kernel_initializer = "he_normal"))

# activation function
cv_model.add(Activation("relu"))

# normalizing batch to improve accuracy of model, normally it is set after convolution layers
cv_model.add(BatchNormalization())

# max pooling layer, for each 2x2 section of input gets the highest pixel value and stores it in a different feature, another way to size down the image
cv_model.add(MaxPooling2D(pool_size = (2, 2)))

# prevents overfitting when neurons get dropped less patterns to remember so it is generalized
cv_model.add(Dropout(0.1))


# convolution layer of 128 convolution filters (updated feature map)
cv_model.add(Conv2D(128, (3, 3), padding = 'same', kernel_initializer = "he_normal"))

# activation to propogate the features
cv_model.add(Activation("relu"))
          
# normalizing batch for accuracy
cv_model.add(BatchNormalization())

# another conv layer of 128
cv_model.add(Conv2D(128, (3, 3), padding = "same", kernel_initializer = "he_normal"))

# activation function
cv_model.add(Activation("relu"))

# normalizing batch for accuracy
cv_model.add(BatchNormalization())

# max pooling layer (remember these layers are essential for optimal performance)
cv_model.add(MaxPooling2D(pool_size = (2, 2)))

# to prevent overfitting
cv_model.add(Dropout(0.1))

# convolution layer of 256 filters (new feature map) 
cv_model.add(Conv2D(256, (3, 3), padding = 'same', kernel_initializer = "he_normal"))

# activation funciton, relu
cv_model.add(Activation("relu"))

# normalizing batch for improvement of accuracy
cv_model.add(BatchNormalization())

# max pooling, extracting max pixel of each 2x2 kernel
cv_model.add(MaxPooling2D(pool_size = (2, 2), padding = "same"))

# another convolution layer this time another 256 conv filters with the same pattern as above
# padding = 'same' allows model to have as much as layers as wanted since input size will always be the same and never shrink to null
cv_model.add(Conv2D(256, (3, 3), padding = 'same', kernel_initializer = "he_normal"))

# activation relu
cv_model.add(Activation("relu"))

# normalizing batch, making features on same scale (like 0-1 for each pixel value)
cv_model.add(BatchNormalization())

# to prevent overfitting deactivate 2% of neurons (less patterns detected, more generalization)
cv_model.add(Dropout(0.1))


# the convolution features become multidimensional, need to make it one dimensional so dense (fully connected) layers do optimal performance as they require one dimensional input
cv_model.add(Flatten())

# 64 neurons layer with l2 regularizer (no mroe l2) of amount 0.0001 (the 0.0001 determines how much of l2)
cv_model.add(Dense(units = 64, kernel_initializer = "he_normal"))

# have to activate the patterns classified to proceed with the next layer utilizign relu funciton
cv_model.add(Activation("relu"))

# normalizing batch to make sure input data on same scale
cv_model.add(BatchNormalization())

# to prevent overfitting so val_loss and val_accuracy have good results
cv_model.add(Dropout(0.1))


# dense, fully connected layers, has l2 regularizer for better performance so no overfitting happens (no ,more)
# l2 regularizers uses its methods to reduce error over val dataset (the 0.0001 determines how much of l2)
# methods such as keeping weights between layers low as possible so model does not pick up too many patterns when training on training data (overfitting)
# mathematical process involving cost function  to do that
cv_model.add(Dense(units=64, kernel_initializer = "he_normal"))

# another activation function of relu
cv_model.add(Activation("relu"))

# normalizing batch to improve accuracy of model
cv_model.add(BatchNormalization())

# prevents overfitting when neurons get dropped less patterns to remember so it is generalized
cv_model.add(Dropout(0.1))

# output layer, going to output 7 outputs since 7 neurons in this layer (each neuron will output a probability for each emotion it thinks it is)
cv_model.add(Dense(units = 7, kernel_initializer = "he_normal", activation = 'softmax'))

# loading weights
cv_model.load_weights("cv_weights.h5")

# loading cv modle
# capturing video frames from default camera (0)
cap = cv2.VideoCapture(0)

# loading training data that will allow for face detection (Haar Cascade)
# this xml document tells classifier how to detect face bassicaly (the guidelines)
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

# we want to transcribe every 200 frames as we do not want the camera to be glitchy (otherwise the processing unit would be running multiple simulatenous operations)
predict_freq = 20

# frame counter to support what the current frame # is
frame_freq = 0


##for index, device in enumerate(PvRecorder.get_audio_devices()):
##    print(index, device)

# setting up audio recorder, recorder var is instance of PvRecorder() object
# frame_length indicates 512 audio samples per frame (remember digital audio course from ap csp) (sampling interval and quanatanization is taken care of)
# recorder = PvRecorder(frame_length = 1000) # the more the value of frame_length the clearer the audio
# audio = [] # saving audio into a list, will be useful when writing audio frames into a wav file

#  # starts recording audio
# recorder.start()

# # using with block to open or create a new file in writing mode
# with wave.open(r"C:\\Users\\coolz\\OneDrive\\Desktop\\AI Research Science Fair Project\\Project\\audio.wav", 'w') as file: # "w" is writing mode with block requires mode for writing functions like writeframes() to work
#     # we are setting the parameters or the protocol for saving the file with setparams()
#     # 1 is the channel meaning mono audio
#     # 2 is the num of bytes per sample (how we manage our digital data)
#     # 16000 samples per second (sample rate)
#     # 1000 audio samples per frame
#     # the two "NONE"s indicate no compression added to the audio (do not want to mix up anything by importing compression algorithms)
#     # first "NONE" indicates lossy or lossless compression, we want none the second one indicates the algorithm for the compression
#     file.setparams((1, 2, 16000, 1000, "NONE", "NONE"))

freq = 44100 # 44100 audio samples per second (utilizing sampling method to capture analog data)
duration = 5 # 5 seconds for each frame of audio

# loading whisper model, small is the type of model (parameters vary for each model) 
model = whisper.load_model("base")

# which audio files have been transcribed
transcribed = []
TRANSCRIPT_FILE = "transcript.txt"

# # try catch to make program very customizeable
# # this try is inside the with block and everything else as each time we are writing file we want to constantly update the audio frames
try:
    # infinite loop since reading from camera/audiio
    while True:
    
        # reading each frame, success is status
        success, image = cap.read()

        # if status is false
        if not success:
            print("Failed to read frame")
            break
        frame_freq += 1  
        
        # tensorflow compatible with rgb
        rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # grayscale for better performance
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    
        # using cascade variable above since it has the ability to now detect (consider it as model variable from tensorflow)
        faces = face_cascade.detectMultiScale(gray, scaleFactor = 1.1, minNeighbors = 8)
        
        # text box
        cv2.rectangle(image, (0, 375), (900, 410), (255, 255, 255), -1)
        cv2.putText(image, "This is a visual to help you be mindful of your emotions", (0, 400), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 0), 2, cv2.LINE_4)
        
        
        face_img = []
        for x, y, w, h in faces:
            # framing it to just the face in grayscale and color
            gray_frame = gray[y:y+h, x:x+w]
            color_frame = image[y:y+h, x:x+w]
                
            # making mutliple rectangles that surround the border of faces detected
            # image is where we make the rectangle, x+w and y+h tells where to put the rectangles (these rectangles form a square)
            # the tuple of 255, 0, 0 is the rgb
            # 4 is the thickness
            cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 4)
                
            # in the frame crop, detect face again to ensure if there is a face
            faces2 = face_cascade.detectMultiScale(gray_frame)
            if len(faces2) == 0:
                print("No face detected")
            else:
                # ex is same as x just had to change name to avoid error
                for (ex, ey, ew, eh) in faces2:
                    # create face crop, changing the coordinates to correspond with the outer square size that detects the face
                    # this is just the frame coordination, needs to be sampled onto an image
                    face_frame = color_frame[ey: ey+h,ex: ex+w]
                    face_frame = cv2.cvtColor(face_frame, cv2.COLOR_BGR2GRAY)
                for face in faces:                
                    # applying face frame to the actual image
                    face_img = cv2.resize(face_frame, (48,48))
            
                    face_img = np.expand_dims(face_img, axis = 0) # need a fourth dimension to enable pooling operations for model
                    # face_img = np.expand_dims(face_img, axis=0)  # add batch dimension

                    face_img = face_img/255.0

                    predictions = cv_model.predict(face_img)


                    # np.argmax will be utilized for the following 7 conditionals as the highest softmax classs is the most likely prediction coming from the model

                    # checks if prediction matches first class
                    if np.argmax(predictions) == 0:
                        cv2.putText(image, "angry", (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_4)
                    elif np.argmax(predictions) == 1:
                        cv2.putText(image, "disgusted", (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_4)
                    elif np.argmax(predictions) == 2:
                        cv2.putText(image, "fearful", (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_4)

                    elif np.argmax(predictions) == 3:
                        cv2.putText(image, "happy", (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_4)
                    elif np.argmax(predictions) == 4:
                        cv2.putText(image, "neutral", (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_4)

                    elif np.argmax(predictions) == 5:
                        cv2.putText(image, "sad", (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_4)
        
                    elif np.argmax(predictions) == 6:
                        cv2.putText(image, "surprised", (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_4)
        
        cv2.imshow("Image", image)
        # reading each audio frame
        # audio_frame = recorder.read()
        
        
        # making sure each audio frame (each 1000 audio samples) is saved to the audio list so that proper contents are written
        # audio.append(audio_frame)

        # "h" is the format specifiying how audio data gets converted to binary (remember quanatanization by converting nums through comm multiples)
        # but here "h" says we only represent the audio samples as short integers (then later converted to binary through the standard way)
        # short ints range from -32,768 to 32,767
        # we do the h format for all the audio samples which is why it is "h" * len(audio), length of audio is # of samples
        # struct.pack() considers "h" to be the format and performs the functionality of converting audio samples to binary so computer understands
        # binary = struct.pack("h" * len(audio_frame), *audio_frame)# the *audio_frame is to make sure that each of 1000 samples are coverted to binary


        # then we use writeframes() to perform the functionality of actually writing the frame recorded (through audio[] since it has all audio data)
        # a single frame since this is live data
        # file.writeframes(binary)

      # # the conditional states when the whisper model can transcribe abiding by the frequency parameters
            # if frame_freq == transcribe_freq:  # transcribe every frequent # of frames to make the camera less laggy and whisper model to be effective
                # using the every frame saved wav file, convert the audio into text, using recgonize()                
                # result = model.transcribe("audio.wav")
                # print(result["text"])



        
        # waiting for a key to pressed in 10ms timeframe
        key = cv2.waitKey(10)

        # ord a function to convert char into a keyboard key
        if key == ord('q'):
            cv2.destroyAllWindows
            cap.release()
            break
# if a key is clicked 
except KeyboardInterrupt:
    print("I hope this was a memorable experience")

```
